```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

from modellib.regression import LinearRegression
```

```python
def example_function(x):
    return (x**2).sum(axis=1) + np.random.randn(x.shape[0])/5
```

```python
x_tr = np.random.uniform(0, 11, (50, 5))
y_tr = example_function(x_tr).reshape(-1, 1)
```

```python
assert x_tr.shape[0] == y_tr.shape[0]
assert x_tr.shape[1] == 5
assert y_tr.shape[1] == 1
```

```python
x_te = np.random.uniform(0, 11, (10, 5))
y_te = example_function(x_te).reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(x_tr[:10], y_tr[:10])
y_ = model.predict(x_te)
```

```python
plt.scatter(y_te, y_.flatten())
plt.suptitle(r2_score(y_te, y_))
```

```python
model.fit(x_tr, y_tr)
y__ = model.predict(x_te)
```

```python
plt.scatter(y_te, y__.flatten())
plt.suptitle(r2_score(y_te, y__))
```
